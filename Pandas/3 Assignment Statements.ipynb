# Assignment Statements
- explains about several techniques for assigning data to cells within a DataFrame.
- usage will be like u get data and you look through the data set, you realize that it's 
necessary to change
small amounts of the data within that data set ie the case Assignment Statements are required

```python
import pandas as pd
import numpy as np
```

```python
starting_date='20160701' # used for creating dates index
sample_numpy_data=np.array(np.arange(24)).reshape(6,4)
dates_index=pd.date_range(starting_date,periods=6)
sample_df=pd.DataFrame(sample_numpy_data,dates_index,columns=list('ABCD'))
sample_df
```

##### 1.SETTING A NEW COLUMN AUTOMATICALLY ALIGNS THE DATA

```python
#lets make a copy
sample_df_2=sample_df.copy()
#lets add a new column in df2
sample_df_2['Fruits']=['apple','orange','bananas','strawberry','blueberry','pineapple']
sample_df_2
#SETTING A NEW COLUMN AUTOMATICALLY ALIGNS THE DATA
```

```python
#Lets create a series
pd.Series([1,2,3,4,5,6], index=pd.date_range(starting_date,periods=6))
```

##### 2.Setting a new column by Data Series

```python
sample_series=pd.Series([1,2,3,4,5,6], index=pd.date_range(starting_date,periods=6))
sample_df_2['Extra Data']=sample_series*3+1
sample_df_2
```

##### 3 setting values by Label

```python
sample_df_2.at[dates_index[3],'Fruits']='Pear'
sample_df_2
```

#### 4.Setting values by Position
- iat 
- documentation - https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.iat.html
- Similarly to iloc, iat provides integer based lookups. You can also set using these indexers    

```python
sample_df_2.iat[3,2]=4444
sample_df_2
```

##### 5.setting by assigning with a numpy array

```python
second_numpy_array=np.array(np.arange(len(sample_df_2))) *100 +7
second_numpy_array
#here sample_df_2 shape is 6,6
#But length is 6 only
```

```python
second_numpy_array=np.array(np.arange(len(sample_df_2))) *100 +7
sample_df_2['G']=second_numpy_array
sample_df_2
```
